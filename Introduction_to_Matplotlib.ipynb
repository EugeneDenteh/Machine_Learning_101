# Introduction to matplotlib
Matplotlib is a Python library for creating static, animated, or interactive visualizations in two and three dimensions. It is highly customizable and versatile, making it a preferred choice for a wide range of data visualization tasks. Matplotlib integrates seamlessly with other Python libraries, offers various plot types, and allows users to finely control the appearance and layout of their visualizations.

# Matplotlib example workflow

```python
#import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#prepare data
x=[1,2,3,4,5]
y=[10,11,14,16,23]

#setup plot
fig,ax=plt.subplots(figsize=(10,10))

#plot data
ax.plot(x,y)

#customize plot
ax.set(title="Example graph",
       xlabel="x-axis",
       ylabel="y-axis")

#save the figure
fig.savefig("example_graph.png")
```

# Making figures with numpy arrays

```python
import numpy as np
```

```python
#create some data
x=np.linspace(0,10,100)
```

```python
#plot the data and create a line plot
#the line plot is the default when you call plot()
fig,ax=plt.subplots()
ax.plot(x,x**2);
```

```python
# Now lets use the same data to create a scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
#another scatter plot
fig, ax=plt.subplots()
#ax.plot(x,np.sin(x));
ax.scatter(x,np.sin(x));
```

```python
#make a plot from a dictionary
nut_butter_prices={"Almond butter":10,"Peanut butter":8,"Cashew butter":12}
fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax.set(title="Nut Butter prices",
     ylabel="Prices ($)");
```

```python
#Create a data using numpy
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

# Creating subplots

```python
#Option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#plot to each axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
#Option 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
heart_disease=pd.read_csv("Data/heart-disease.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist(bins=100);
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True,bins=20);
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50
```

```python
# pyplot method
over_50.plot(kind="scatter",x="age",y="chol",c="target");
```

```python
# Object oriented method
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
#ax.set_xlim([45,100]) #this changes the limits of the x-axis
```

```python
# OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))

# plot the data
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);

# customize the plot
ax.set(title="Heart Disease and Cholestrol levels",xlabel="Age",ylabel="Cholesterol");

# add a legend
ax.legend(*scatter.legend_elements(),title="Target");
# add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
#subplot of chol, age and thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))
# add data to ax0
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
# customize ax0
ax0.set(title="Heart Disease and Cholesterol levels",xlabel="Age",ylabel="Cholesterol")
#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")
#add a mean line
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

# add data to ax1
scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])
# customize ax1
ax.set(title="Heart Disease and Max Heart Rate",xlabel="Age",ylabel="Max Heart Rate");
# add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")
# add a meanline
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");
# add a title to the figure
fig.suptitle("Heart Disease analysis",fontsize=16,fontweight="bold");
```
